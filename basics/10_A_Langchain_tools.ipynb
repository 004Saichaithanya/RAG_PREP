Custom Tools

Method 1 : using @tool decorator

```python
from langchain_core.tools import tool

@tool
def multiply(a : int , b : int) -> int:
    """Multiply two numbers"""
    return a*b

result = multiply.invoke({'a':3,'b':2})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
multiply.args_schema.model_json_schema()
```

Method 2 : using Structured Tool & Pydantic 

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):
    a : int = Field(required = True , description="the first number to multiply")
    b : int = Field(required = True , description="the second number to multiply")
```

```python
def multiply_func(a : int ,b : int) -> int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = "mulitply",
    description= "Mulitply two number using StructuredTool",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3,'b':2})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
multiply_tool.args_schema.model_json_schema()
```

Method 3 : using Base Tool

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# pydantic model 
class MultiplyInput(BaseModel):
    a : int = Field(required = True , description="the first number to multiply")
    b : int = Field(required = True , description="the second number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name: str  = "multiply"
    description:str = "multiply two number using BaseTool"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self,a : int , b: int) -> int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':5,'b':10})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
multiply_tool.args_schema.model_json_schema()
```

ToolKit

```python
@tool
def add(a : int , b: int) -> int:
    """Add two numbers"""
    return a+b

@tool
def multiply(a : int , b : int)-> int:
    """Multiply two numbers"""
    return a*b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add,multiply]
```

```python
toolkit = MathToolKit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name , '->',tool.description)
```

```python
print(tools[0].invoke({'a':10,"b":5}))
print(tools[1].invoke({'a':10,"b":5}))
```

Tool Calling

```python
from langchain_core.tools import tool
from langchain_openrouter import ChatOpenRouter
```

```python
@tool
def add(a : int , b: int) -> int:
    """Add two numbers"""
    return a+b

@tool
def multiply(a : int , b : int)-> int:
    """Multiply two numbers"""
    return a*b


@tool
def rev_string(s : str)-> str:
    """Reverse the String"""
    return s[::-1]
```

```python
class LLMTools:
    def get_tools(self):
        return [add,multiply,rev_string]
```

```python
llm_tools = LLMTools()
```

```python
tools = llm_tools.get_tools()
```

```python
llm = ChatOpenRouter(
    model = 'qwen/qwen3-32b'
)
llm
```

```python
llm_with_tools = llm.bind_tools(tools)
```

```python
input_1 = "what is the addition of 10 and 20"

output = llm_with_tools.invoke(input_1)
```

```python
print(output)
```

```python
print(output.content)
```

```python
print(output.tool_calls)
```

```python
response = llm_with_tools.invoke(
    "what is the addition of 10 and 20"
)

print(response.tool_calls)
```

```python
tool_call = response.tool_calls[0]

result = add.invoke(tool_call["args"])

print(result)
```

```python
from langchain_core.messages import ToolMessage,HumanMessage
```

```python
tool_msg = ToolMessage(
    content=str(result),
    tool_call_id=tool_call["id"]
)

final_response = llm_with_tools.invoke(
    [response, tool_msg]
)

print(final_response.content)
```

Sending Everything to LLM for results 

```python
from langchain_core.messages import ToolMessage,HumanMessage
```

```python
query = HumanMessage(input('enter the query'))
messages = [query]
messages
```

```python
response = llm_with_tools.invoke(messages)
messages.append(response)
messages
```

```python
tools = [add,multiply,rev_string]
```

```python
tools_map={
    tool.name:tool for tool in tools
}
tools_map.keys()
```

```python
tool_call = response.tool_calls[0]
tool_call
```

```python
tool_msg = tools_map[tool_call['name']].invoke(response.tool_calls[0])
tool_msg
```

```python
messages.append(tool_msg)
messages
```

```python
final_response = llm_with_tools.invoke(messages).content
final_response
```
